## Step 3 : Exploratory Data Analysis (EDA) â€” Inspect, Visualize and Clean the Data

#### Above, we can see that the column headers need some cleaning as well as the values in the data. 

From the UCI webpage, there is a names file. Opening that as a text file, we can pull some information and construct the below dict. This dict will be used to decode data in the original file. 

From there, we can remove or interpolate any missing data to create our cleaned data set. 

```python
col_name_value_dict={
    'attribute':{'poisonous':'p', 'edible':'e'},
    'cap_shape':{'bell':'b','conical':'c','convex':'x','flat':'f','knobbed':'k','sunken':'s'},
    'cap_surface':{'fibrous':'f','grooves':'g','scaly':'y','smooth':'s'},
    'cap_color':{'brown':'n','buff':'b','cinnamon':'c','gray':'g','green':'r','pink':'p','purple':'u','red':'e','white':'w','yellow':'y'},
    'has_bruises':{'bruises':'t','no':'f'},
    'odor':{'almond':'a','anise':'l','creosote':'c','fishy':'y','foul':'f','musty':'m','none':'n','pungent':'p','spicy':'s'},
    'gill_attachment':{'attached':'a','descending':'d','free':'f','notched':'n'},
    'gill_spacing':{'close':'c','crowded':'w','distant':'d'},
    'gill_size':{'broad':'b','narrow':'n'},
    'gill_color':{'black':'k','brown':'n','buff':'b','chocolate':'h','gray':'g','green':'r','orange':'o','pink':'p','purple':'u','red':'e','white':'w','yellow':'y'},
    'stalk_shape':{'enlarging':'e','tapering':'t'},
    'stalk_root':{'bulbous':'b','club':'c','cup':'u','equal':'e','rhizomorphs':'z','rooted':'r','missing':'?'},
    'stalk_surface_above_ring':{'fibrous':'f','scaly':'y','silky':'k','smooth':'s'},
    'stalk_surface_below_ring':{'fibrous':'f','scaly':'y','silky':'k','smooth':'s'},
    'stalk_color_above_ring':{'brown':'n','buff':'b','cinnamon':'c','gray':'g','orange':'o','pink':'p','red':'e','white':'w','yellow':'y'},
    'stalk_color_below_ring':{'brown':'n','buff':'b','cinnamon':'c','gray':'g','orange':'o','pink':'p','red':'e','white':'w','yellow':'y'},
    'veil_type':{'partial':'p','universal':'u'},
    'veil_color':{'brown':'n','orange':'o','white':'w','yellow':'y'},
    'ring_number':{'none':'n','one':'o','two':'t'},
    'ring_type':{'cobwebby':'c','evanescent':'e','flaring':'f','large':'l','none':'n','pendant':'p','sheathing':'s','zone':'z'},
    'spore_print_color':{'black':'k','brown':'n','buff':'b','chocolate':'h','green':'r','orange':'o','purple':'u','white':'w','yellow':'y'},
    'population':{'abundant':'a','clustered':'c','numerous':'n','scattered':'s','several':'v','solitary':'y'},
    'habitat':{'grasses':'g','leaves':'l','meadows':'m','paths':'p','urban':'u','waste':'w','woods':'d'},
}
```

```python
raw_data.columns = list(col_name_value_dict.keys())
```

```python
# let's take another look at our data with our now readable headers
raw_data.head()
```

```python
# We also see that the stalk_root feature has "?" in place of missing values. Per the data docs there are ~2k of these. 
# We should still have enough data if we remove these rows


# we will also drop any rows with missing data
initial_data = raw_data[raw_data['stalk_root']!="?"].dropna()
```

```python
print(f"""{len(initial_data)} rows of mushroom data loaded remain. """)
```

```python
initial_data.dtypes
```

#### Great! Our final data cleaning step is to transform our table of string categorical variables into something more computationally consumable - numbers. 

To do this, we will utilize a methodology called "One Hot Encoding." Conveniently, there is a sklearn function for this exact purpose. 

Note: Because our features do not have order, we cannot use orinal encoding.

```python
# let's create our list of categories and calculate the expected number of resulting columns after encoding
# since the encoding does not name out columns for us based on the previous name and the category, we need to create this manually.

import numpy as np

cat_codes = []
cat_names_all = []
cat_cnt = 0

for col in initial_data:

    cat_cnt += len(col_name_value_dict[col].values())
    cat_codes.append(list(col_name_value_dict[col].values()))
    for cat in col_name_value_dict[col].keys():
            cat_names_all.append(col+'-'+cat)


print(f"""{cat_cnt} columns expected """)
```

```python
# perform the encoding
from sklearn.preprocessing import OneHotEncoder
enc = OneHotEncoder(handle_unknown='error', categories=cat_codes)
sparce_initial_data = enc.fit_transform(initial_data)
```

```python
# get our data back into a dataframe for visual simplicity (this is not really needed from a modelling perspective).

import scipy.sparse

transformed_df = pd.DataFrame.sparse.from_spmatrix(sparce_initial_data).sparse.to_dense()
transformed_df.columns = cat_names_all
print(f"""transformed df has {transformed_df.shape[1]} columns""")


# last in the transormation, we have to drop our extra target variable. we will keep the col that has a 1 for if the mushroom was poisonous
reduced_transformed_df = transformed_df.drop(columns=['attribute-edible']).rename(columns={'attribute-poisonous':'is_poisonous'})

print(f"""reduced transformed df has {reduced_transformed_df.shape[1]} columns""")
```

```python
reduced_transformed_df.head()
```

## Cleaning Conclusions

### Because the data is 100% categorical variables / features, we needed to perform a One-Hot encoding to convert the string data into numeric represented boolean data. We were able to do this with the help of an sklearn package and we can now begin to vizualize. 

### Also - due to the nature of 100% categorical data, our previously condensed dataset is now pretty sparse in nature (but has been converted technically).

#### Now that the data is cleaned, we should split it BEFORE we start to vizualize and play around with explanation. 

```python
from sklearn.model_selection import train_test_split

# let's keep the standard 20% as our testing data
train_df, test_df = train_test_split(reduced_transformed_df, test_size=0.25)
```

# Step 4 : Perform Analysis Using Supervised Machine Learning Models

```python
# first here we will need to split our X and y : 
x_train = train_df.drop(columns=['is_poisonous'])
y_train = train_df[['is_poisonous']]
x_test = test_df.drop(columns=['is_poisonous'])
y_test = test_df[['is_poisonous']]
```

### Now we will run multiple classification models to see which one offers the most accurate prediction

For this project, we will create and compare 4 models:
1. Decisions Tree Classification
2. Adaboost Classification
3. Random Forest Classification
4. Support Vector Machine


```python
from sklearn import tree

dt = tree.DecisionTreeClassifier()
dt = dt.fit(x_train, np.array(y_train).reshape(-1,))
dt_score_train = dt.score(x_train, y_train)
dt_score_test = dt.score(x_test, y_test)

print("Basic Decision Tree model (no tuning) has a default training score or {:.2%} and a testing score of {:.2%}. ".format(dt_score_train,dt_score_test))
```

```python
# 

from sklearn.ensemble import AdaBoostClassifier

ab = AdaBoostClassifier()
ab = ab.fit(x_train, np.array(y_train).reshape(-1,))
ab_score_train = ab.score(x_train, y_train)
ab_score_test = ab.score(x_test, y_test)

print("Basic Adaboost model (no tuning) has a default training score or {:.2%} and a testing score of {:.2%}. ".format(ab_score_train,ab_score_test))
```

```python
# 

from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier()
rf = rf.fit(x_train, np.array(y_train).reshape(-1,))
rf_score_train = rf.score(x_train, y_train)
rf_score_test = rf.score(x_test, y_test)

print("Basic Random Forest model (no tuning) has a default training score or {:.2%} and a testing score of {:.2%}. ".format(rf_score_train,rf_score_test))
```

```python
# this model was not covered in class

from sklearn.ensemble import GradientBoostingClassifier

gb = GradientBoostingClassifier()
gb = gb.fit(x_train, np.array(y_train).reshape(-1,))
gb_score_train = gb.score(x_train, y_train)
gb_score_test = gb.score(x_test, y_test)

print("Basic Gradient Boosting model (no tuning) has a default training score or {:.2%} and a testing score of {:.2%}. ".format(gb_score_train,gb_score_test))
```

## Model Conclusion:

### And here we are, all of our basic tree models seem to do the trick! So this is good; if we can find enough mushrooms to eat we should be 100% confident that we won't poison ourselves by using any of these models. 
Since this score is already as high as we can hope, we do not need to tune it.
