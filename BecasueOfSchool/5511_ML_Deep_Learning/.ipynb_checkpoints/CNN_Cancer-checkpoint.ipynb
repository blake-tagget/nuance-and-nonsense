### Model Architecture 

Now that we have our data imported and a handle on the number of images we can utilize, let's build a basic CNN model. 

Generally speaking, we will want to make a cake. Since this is "deep learning" we are going to add a bunch of layers to our cake. Some or most layers will be convolutional, while others will be other types like (Max Pooling, Fully connected / Dense, Dropout, etc). And like all cakes we will have a base layer (input) and a top layer (output). Since the job we are doing is binary classification, out output later will produce a single value - the probability that the image contains metastasized cancerous cells. 



Arch choices:

VGGNet - n * (2 Conv + 1 Pool) + 7 dense + softmax

Make it really deep ... 


```python

limit = 25
limit = None

batch_size = 32
img_pro = tf.keras.preprocessing.image.ImageDataGenerator(rescale=1 / 255.0, validation_split=0.10)

train_img = img_pro.flow_from_dataframe(
    train_labels.head(limit),
    directory=train_folder,
    x_col='file_name',
    y_col='label_string',
    color_mode='rgb',
    class_mode='binary',
    batch_size=batch_size,
    shuffle=False,
    subset="training"
)

valid_img = img_pro.flow_from_dataframe(
    train_labels.head(limit),
    directory=train_folder,
    x_col='file_name',
    y_col='label_string',
    color_mode='rgb',
    class_mode='binary',
    batch_size=batch_size,
    shuffle=False,
    subset="validation"
)


test_img = img_pro.flow_from_dataframe(
    test_ids.head(limit),
    directory=test_folder,
    x_col='file_name',
    color_mode='rgb',
    class_mode=None,
    batch_size=batch_size,
    shuffle=False,
)
```

```python
model = tf.keras.Sequential([
  tf.keras.layers.Conv2D(32, 3, activation='relu'),
  tf.keras.layers.Conv2D(32, 3, activation='relu'),
  tf.keras.layers.MaxPooling2D(),
  tf.keras.layers.Conv2D(32, 3, activation='relu'),
  tf.keras.layers.Conv2D(32, 3, activation='relu'),
  tf.keras.layers.MaxPooling2D(),
  tf.keras.layers.Flatten(),
  tf.keras.layers.Dense(128, activation='relu'),
  tf.keras.layers.Dense(1, activation='sigmoid') 
])

model.compile(
  optimizer='adam',
  loss=tf.keras.losses.BinaryCrossentropy(),
  metrics=['accuracy'])
```

```python
stamp('start')
model.fit(
  train_img,
  validation_data=valid_img,
  epochs=3
)
stamp('end')
```

```python
stamp('start')
test_copy = test_ids[['id']].head(limit)
test_copy['label'] = (model.predict(test_img)>.5)*1
stamp('end')
test_copy.head()
```

```python
test_copy.to_csv(f'''{gdrive_path}submission_results_{time.strftime("%Y.%m.%d.%H:%M:%S", time.localtime())}.csv''')
```
